```python
# 프로펫은 2개 컬럼만 있으면, 예측 가능하다.
# 날짜 컬럼과, 예측하려는 컬럼!!!
```

```python
df
```

```python
df_avocado = df[ ['Date', 'AveragePrice'] ]
```

```python
df_avocado
```

```python
# 1. 컬럼명을, 프로펫 라이브러리가 인식할 수 있도록 바꿔준다.
```

```python
df_prophet = df_avocado.rename( columns={'Date':'ds' , 'AveragePrice':'y'} )
```

```python
df_prophet
```

```python
# 2. 데이터를 학습시킨다.
```

```python
from prophet import Prophet
```

```python
m = Prophet()
```

```python
m.fit(df_prophet)
```

```python
# 3. 예측에 사용할 데이터 프레임 만든다.
```

```python
future = m.make_future_dataframe(periods= 365, freq= 'D' )
```

```python
# 4. 예측한다.
```

```python
forecast = m.predict(future)
```

```python
forecast
```

```python
# 5. 예측 결과를 차트로 나타낸다.
```

```python
m.plot(forecast, xlabel='Date', ylabel='Price')
plt.savefig('fig1.jpg')
```

```python
m.plot_components(forecast)
plt.savefig('fig2.jpg')
```
